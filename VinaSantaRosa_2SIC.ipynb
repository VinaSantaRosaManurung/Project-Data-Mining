# Merubah type data

```python
def make_label(nama_kabupaten_kota):
    if nama_kabupaten_kota == "Jakarta Utara":
        return 1
    elif nama_kabupaten_kota == "Jakarta Selatan":
        return 2
    else:
        return 0
```

```python
data.head()
```

```python
data["Nama Kabupaten Kota"]=data["nama_kabupaten_kota"].apply(make_label)
data
```

```python
def make_label(nama_destinasi_wisata):
    if nama_destinasi_wisata == "Planetarium":
        return 1
    elif nama_destinasi_wisata == "Museum Prasasti":
        return 2
    else:
        return 0
```

```python
data["Nama Destinasi Wisata"]=data["nama_destinasi_wisata"].apply(make_label)
data
```

```python
def make_label(nama_provinsi):
    if nama_provinsi == "DKI Jakarta":
        return 1
    elif nama_provinsi == "":
        return 2
    else:
        return 0
```

```python
data["Nama Provinsi"]=data["nama_provinsi"].apply(make_label)
data
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(data[["Nama Destinasi Wisata","Nama Provinsi","jumlah"]],data["Nama Kabupaten Kota"])
```

# BernoulliNB

```python
from sklearn import naive_bayes
```

```python
clf = naive_bayes.BernoulliNB()
```

```python
clf.fit(x_train, y_train)
```

```python
clf.predict(x_test)
```

```python
clf.score(x_train,y_train) # data training = 80%
```

```python
clf.score(x_test,y_test) #data testing = 20%
```

# GaussianNB

```python
from sklearn.naive_bayes import GaussianNB
```

```python
modelnb = GaussianNB()
```

```python
nbtrain = modelnb.fit(x_train,y_train)  # data training 80%
nbtrain.class_count_
```

```python
modelnb.predict(x_test)
```

```python
modelnb.score(x_train,y_train) # data training 80%
```

```python
modelnb.score(x_test,y_test) # data testing 20%
```
